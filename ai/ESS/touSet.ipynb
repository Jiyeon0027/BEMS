```python
import pandas as pd
```

```python
df = pd.read_csv("../../preprocessed_data/[1분 단위]아파트_동별_소비전력_전력분배_2022-07-18 00.00.00~2023-08-30 10.39.00.csv", usecols=[i for i in range(1,5)])
df.info()
```

```python
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])

df.info()
```

```python
df['TOTAL_CONSUMPTION(kWh)'] = (df['561_CONSUMPTION(kW)'] + df['562_CONSUMPTION(kW)'] + df['563_CONSUMPTION(kW)']) * 60
df
```

```python
df.drop(['561_CONSUMPTION(kW)', '562_CONSUMPTION(kW)', '563_CONSUMPTION(kW)'], axis=1, inplace = True)
df
```

```python
df.head(3).to_dict()
```

```python
# df['TOTAL_CONSUMPTION(MWh)'] = df['TOTAL_CONSUMPTION(kWh)'] / 1000
```

```python
summer_start = "2023-06-01"
summer_end = "2023-09-01"

summer_df = df.query("`TIMESTAMP` >= @summer_start and `TIMESTAMP` <= @summer_end")

summer_df
```

```python
int(len(summer_df) * 0.99)
```

```python
summer_train = summer_df.iloc[:int(len(summer_df) * 0.99)]
summer_valid = summer_df.iloc[int(len(summer_df) * 0.99):]

print(f"train len {len(summer_train)}, valid len {len(summer_valid)}")
```
